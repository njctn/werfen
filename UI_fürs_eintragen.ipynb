```python
import tkinter as tk
from tkinter import messagebox
from openpyxl import Workbook, load_workbook
import random
import pandas as pd

# SET BEFORE STARTING
#num_teams = 4

# reset the global var teams so that no leftover data gets insorted
teams = None
teams_df = None
num_teams = None

class Teams:
    def __init__(self, master):
        self.master = master
        self.master.title("Teams")
        
        #self.labels = []
        self.teams = []
        
        # Label T1 P1
        self.label = tk.Label(master, text="Player 1")
        self.label.grid(row=0, column=0)
        # field T1 P1
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=0, column=1)
        self.teams.append(self.data_entry)
        
        # Label T1 P2
        self.label = tk.Label(master, text="Player 2")
        self.label.grid(row=1, column=0)
        # field T1 P2
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=1, column=1)
        self.teams.append(self.data_entry)
        
        # Label T1 P3
        self.label = tk.Label(master, text="Player 3")
        self.label.grid(row=2, column=0)
        # field T1 P1
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=2, column=1)
        self.teams.append(self.data_entry)
        
        # Label T1 
        self.label = tk.Label(master, text="Teamname 1")
        self.label.grid(row=3, column=0)
        # field T1
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=3, column=1)
        self.teams.append(self.data_entry)
        
        # Label T2 P1
        self.label = tk.Label(master, text="Player 4")
        self.label.grid(row=0, column=3)
        # field T2 P1
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=0, column=4)
        self.teams.append(self.data_entry)
        
        # label T2 P2
        self.label = tk.Label(master, text="Player 5")
        self.label.grid(row=1, column=3)
        # field T2 P2
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=1, column=4)
        self.teams.append(self.data_entry)
        
        # label T2 P3
        self.label = tk.Label(master, text="Player 6")
        self.label.grid(row=2, column=3)
        # field T2 P3
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=2, column=4)
        self.teams.append(self.data_entry)
        # label T2 
        self.label = tk.Label(master, text="Teamname 2")
        self.label.grid(row=3, column=3)
        # field T2
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=3, column=4)
        self.teams.append(self.data_entry)
        
        # label T3 P1
        self.label = tk.Label(master, text="Player 7")
        self.label.grid(row=0, column=5)
        # field T3 P1
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=0, column=6)
        self.teams.append(self.data_entry)
        
        # label T3 P2
        self.label = tk.Label(master, text="Player 8")
        self.label.grid(row=1, column=5)
        # field T3 P2
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=1, column=6)
        self.teams.append(self.data_entry)
        
        # field T3 P3
        self.label = tk.Label(master, text="Player 9")
        self.label.grid(row=2, column=5)
        # label T3 P3
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=2, column=6)
        self.teams.append(self.data_entry)
        
        # label T3
        self.label = tk.Label(master, text="Teamname 3")
        self.label.grid(row=3, column=5)
        # field T3
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=3, column=6)
        self.teams.append(self.data_entry)   
        
        # label T4 P1
        self.label = tk.Label(master, text="Player 10")
        self.label.grid(row=0, column=7)
        # field T4 P1
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=0, column=8)
        self.teams.append(self.data_entry)
        
        # label T4 P2
        self.label = tk.Label(master, text="Player 11")
        self.label.grid(row=1, column=7)
        # field T4 P2
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=1, column=8)
        self.teams.append(self.data_entry)
        
        # field T4 P3
        self.label = tk.Label(master, text="Player 12")
        self.label.grid(row=2, column=7)
        # label T4 P3
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=2, column=8)
        self.teams.append(self.data_entry)
        
        # label T4
        self.label = tk.Label(master, text="Teamname 4")
        self.label.grid(row=3, column=7)
        # field T4
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=3, column=8)
        self.teams.append(self.data_entry) 
        
        self.insert_button = tk.Button(master, text="Insert", command=self.insert_data)
        self.insert_button.grid(row=10, column=0, columnspan=1)
        
        # Button that generates random data, only for testing
        self.random_button = tk.Button(master, text="Random", command=self.generate_random_data)
        self.random_button.grid(row=10, column=1, columnspan=1)
        
        self.save_button = tk.Button(master, text="Save", command=self.save_data)
        self.save_button.grid(row=10, column=2, columnspan=1)

    def insert_data(self):
        global teams
        teams = [field.get() for field in self.teams]
        

        # Check if data is not empty
        if all(teams):
            try:
                # Open or create Excel files
                spiele = load_workbook("Spiele.xlsx")
                #workbook2 = load_workbook("excel_file2.xlsx")

                # Accessing sheets in Excel files
                sheet1 = spiele.active
                #sheet2 = workbook2.active

                # Inserting data into Excel sheets
                sheet1.append(teams[:])
                #sheet2.append([data])

                # Saving changes to Excel files
                spiele.save("Spiele.xlsx")
                #workbook2.save("excel_file2.xlsx")

                messagebox.showinfo("Success", "Data inserted successfully into Excel sheets!")
            except Exception as e:
                messagebox.showerror("Error", f"An error occurred: {str(e)}")
        else:
            messagebox.showwarning("Warning", "Please enter some data!")
            
    def get_data(self):
        return self.teams
    
    def save_data(self):
        global teams, teams_df, num_teams
        df_cols = ['Player 1', 'Player 2', 'Player 3', 'Teamname']
        teams = [field.get() for field in self.teams]
        teams_df = [teams[i:i+4] for i in range(0, 16, 4)]
        teams_df = pd.DataFrame(teams_df, columns = df_cols)
        
        if (teams_df.iloc[3,0] == teams_df.iloc[3,1] == teams_df.iloc[3,2] == teams_df.iloc[3,3] == ''):
            num_teams = 3
        
    
    def generate_random_data(self):
        random_data = [str(random.randint(1, 100)) for _ in range(16)] # Generating random integers for each column
        for field, value in zip(self.teams, random_data):
            field.delete(0, tk.END)  # Clear previous data
            field.insert(0, value)    # Insert random data into entry field

            
class Matches:
    def __init__(self, master):
        self.master = master
        self.master.title("Matches")
        
        self.data = []
         
            
        # Label T1 P1
        self.label = tk.Label(master, text="Player 1")
        self.label.grid(row=0, column=0)
        # field T1 P1
        self.data_entry = tk.Entry(master)
        self.data_entry.grid(row=0, column=1)
        self.teams.append(self.data_entry)
        
        # List to hold the Checkbutton variables
        self.selection = []
        
        # Create Checkbuttons
        options = ["Option 1", "Option 2", "Option 3"]
        for i, option in enumerate(options):
            var = tk.IntVar()
            cb = tk.Checkbutton(master, text=option, variable=var, command=self.validate_selection)
            cb.grid(row=i, column=0, sticky="w")
            self.selection.append(var)
            
    def validate_selection(self):
        selected_count = sum(var.get() for var in self.selection_vars)
        if selected_count > 2:
            messagebox.showwarning("Warning", "Please select only 2 options.")
            # Unselect the last checked option
            for var in self.selection_vars:
                if var.get() == 1:
                    var.set(0)
        

def on_closing():
    root.destroy()
    
def main():
    global root
    root = tk.Tk()
    app = Teams(root)
    root.protocol("WM_DELETE_WINDOW", on_closing)  # Set the closing event handler
    root.mainloop()

if __name__ == "__main__":
    main()
```
